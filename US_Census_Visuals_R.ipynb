# R Visuals in ggplot 

Let's start with the U.S. Census Bureau population projections.

```r
library(ggplot2)
library(RColorBrewer)

pproj = read.csv("../../../datasets/census/pop_proj.csv")
head(pproj)
summary(pproj)
```

This data set has seven age brackets instead of each and every age. Years show the projection years of the male and female population in U.S. and population is given in 1000s. 

Here's bar plot to show the age bracket distribution of the total population in year 2015. 

```r
# basic bar chart
ggplot(pproj, aes(x=age, y=pop)) +
       geom_bar(data = subset(pproj, year==2015), stat='identity') +
       scale_x_discrete(limits = c("Under 5 years","5 to 13 years","14 to 17 years","18 to 24 years","25 to 44 years", 
                            "45 to 64 years","65 years and over"))
```

Now we'll create a pie chart 

```r
summary(pproj)
str(pproj)
```

```r
ggplot(pproj, aes(x=factor(1), y=pop, fill=age, color=age)) + 
geom_bar(data = subset(pproj, year==2015), stat="identity", width=1) +
coord_polar(theta="y")
```

Now I'll create a pie chart for similar to above exercise but only for female population. 

```r
ggplot(pproj, aes(x=factor(1), y=pop, fill=age, color=age)) + 
geom_bar(data = subset(pproj, year==2015 & sex=="Female"), stat="identity", width=1) +
coord_polar(theta="y")
```

Now I'll create pie charts of all years for total population similar to exercise 2. Use faceting.

```r
ggplot(pproj, aes(x=factor(1), y=pop, fill=age, color=age)) + 
geom_bar(stat="identity", width=1, position='fill') +
coord_polar(theta="y") +
facet_grid(.~year) 
```

Here's the steps needed to create a population pyramid for year 2015. 

```r
ggplot(pproj, aes(x=age,y=pop, fill=sex, color=sex)) + 
geom_bar(data = subset(pproj, year==2015), stat = "identity") +
geom_bar(data = subset(pproj, sex=="Female"), stat = "identity") +
geom_bar(data = subset(pproj, sex=="Male"), stat = "identity", position = "identity", aes(y=-pop)) +
coord_flip() +
theme_minimal() +
scale_y_continuous(labels = abs) +
ggtitle("U.S. Population (2015 Census)")
 
```

Now create four population pyramids for all years. Use faceting.  

```r
# Look at your practice code and figure out how you could adapt it for this. 
ggplot(pproj, aes(x=age,y=pop, fill=sex, color=sex)) + 
geom_bar(data = subset(pproj, sex=="Female"), stat = "identity") +
geom_bar(data = subset(pproj, sex=="Male"), stat = "identity", position = "identity", aes(y=-pop)) +
facet_grid(.~year)
coord_flip() +
theme_minimal() +
scale_y_continuous(labels = abs) +
scale_x_discrete(limits = c("Under 5 years","5 to 13 years","14 to 17 years","18 to 24 years","25 to 44 years", 
                            "45 to 64 years","65 years and over")) +
ggtitle("U.S. Population (2015 Census)")
 
```

Now we will read another data set: San Francisco population projections from California Open Data Portal. This data has population projections for each age and ethnicity. 

Now I'll create a **total** population pyramid for year 2015.

```r
sfoproj = read.csv("../../../datasets/census/SFO_pop_pyr.csv")
head(sfoproj)
summary(sfoproj)
```

```r
ggplot(sfoproj, aes(x=Age,y=Population, fill=Gender)) + 
geom_bar(data = subset(sfoproj, Year==2015 & Gender=="Female"), stat = "identity") +
geom_bar(data = subset(sfoproj, Year==2015 & Gender=="Male"), stat = "identity", position = "identity", aes(y=-Population)) +
coord_flip() +
theme_minimal() +
scale_y_continuous(labels = abs) +
ggtitle("U.S. Population (2015 Census)")
```

Here's a population pyramid for year 2015, while showing ethnicities as stacked bars. 

```r
# First we need to create an "ordered" attribute so that both sides 
# of the pyramid are ordered with respect to the order we specify. 
sfoproj <- transform(sfoproj,
       Race_name = ordered(Race_name, levels = c("White","Asian","Hispanic","Black","American Indian","Pacific Islander","Multi-Race")))

# Now do the pyramid here. do NOT use  position = "identity" for the geom_bar 
ggplot(sfoproj, aes(x=Age, y=Population, fill=Race_name)) + 
geom_bar(data = subset(sfoproj, Year==2015 & Gender=='Female'), stat = 'identity', width=0.7) +
geom_bar(data = subset(sfoproj, Year==2015 & Gender=='Male'), width=0.7, stat = "identity", aes(y=-Population)) +
coord_flip() +
theme_minimal() +
scale_y_continuous(labels = abs) +scale_fill_brewer(palette="Set2")
```
